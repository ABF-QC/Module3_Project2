```python
import requests
import xml.etree.ElementTree as ET

# API URL
url = "https://boardgamegeek.com/xmlapi2/thing?id=169786&comments=1"

# Request response
response = requests.get(url)

if response.status_code == 200:
    root = ET.fromstring(response.content)

    # Extracting game name
    name = root.find(".//name[@type='primary']").attrib["value"]
    print(f"Game: {name}")

    # Extracting first 10 comments
    comments = root.findall(".//comment")[:10]
    
    for idx, comment in enumerate(comments, 1):
        user = comment.attrib["username"]
        rating = comment.attrib.get("rating", "N/A")
        text = comment.attrib["value"]
        print(f"\n{idx}. {user} (Rating: {rating}): {text[:100]}...")
else:
    print("Failed to fetch data")
```

```python
import requests
import xml.etree.ElementTree as ET
import csv

# API URL
url = "https://boardgamegeek.com/xmlapi2/thing?id=169786&comments=1"

# Request response
response = requests.get(url)

if response.status_code == 200:
    root = ET.fromstring(response.content)

    # Extracting game name
    name = root.find(".//name[@type='primary']").attrib["value"]
    print(f"Game: {name}")

    # Prepare CSV file
    with open('BG_comments.csv', mode='w', newline='', encoding='utf-8') as file:  # Changed filename here
        writer = csv.writer(file)
        
        # Write header
        writer.writerow(["Username", "Rating", "Comment"])
        
        # Extracting all comments
        comments = root.findall(".//comment")
        
        for idx, comment in enumerate(comments, 1):
            user = comment.attrib["username"]
            rating = comment.attrib.get("rating", "N/A")
            text = comment.attrib["value"]
            
            # Write each comment to CSV file
            writer.writerow([user, rating, text])
            print(f"\n{idx}. {user} (Rating: {rating}): {text[:100]}...")

    print(f"Data saved to BG_comments.csv")
else:
    print("Failed to fetch data")
```
